# 결측값 처리 부분 볷습필수
### 데이터전처리: 결측값처리?

통계/집계함수

```python
import pandas as pd

data = {
    "상품명": ["무선 이어폰", "스마트 워치", "텀블러", "노트북", "블루투스 스피커", "무드등"],
    "가격": [129000, 250000, 15000, 1200000, 85000, 22000],
    "재고": [23, 12, 54, 5, 17, 31]
}
df = pd.DataFrame(data)
print(df)
```

```python
print(f"디스크라이브:\n{df.describe()}")
print(f"디스크라이브:\n{df["가격"].describe()}")

print(f"평균: {df["가격"].mean()}")
print(f"중앙값: {df["재고"].median()}")
print(f"표준편차: {df["가격"].std()}")
print(f"분산: {df["재고"].var()}")
print(f"값개수: {df["상품명"].count()}")
print(f"최대값: {df["가격"].max()}")
print(f"최솟값: {df["재고"].min()}")
print(f"전체개수: {df["재고"].sum()}")
```

```python
print("최대값 위치 찍기\n",df["가격"].idxmax())
print("최소값 위치 찍기\n",df["가격"].idxmin())

#Nan: Not A Number

df.describe(include="all")
# Nan, 결측값입니다.
```

```python
import numpy as np

data = {
    "이름": ["서준", "하은", "민준", "서연", np.nan, "지민"],
    "나이": [22, 28, np.nan, 31, 27, 24],
    "점수": [89, np.nan, 83, 90, 88, 93]
}
df = pd.DataFrame(data)

# 결측값 탐지
print(df.isnull())
# 결측값이 맞으면 True 아니면 False반환

# 결측값이 어디에 총 몇개 있나?
print(df.isnull().sum())
# 그래서 이 데이터에서 총 몇개?
print(df.isnull().sum().sum())
```

```python
# notnull
df.notnull()
# isnull과 반대

data = {
    "이름": ["서준", "하은", "민준", "서연", np.nan, "지민"],
    "나이": [22, 28, np.nan, 31, 27, 24],
    "점수": [89, 39, 83, 90, 88, 93]
}
df = pd.DataFrame(data)

# 결측값이 존재하는 행을 아예 삭제해버림
df_drop=df.dropna()
print(df_drop,"\n")

df3=df.dropna(axis=1) #열을 지워버린다.
print(df3)

df4=df.fillna(0)
print(df4) # na 값을 다 0으로 변경

avg_age=df["나이"].mean()
df5=df["나이"].fillna(avg_age)
print("\n",df5) # s나이를 평균으로 대체 

df6=df.fillna(method="ffill")
print(df5)

df6=df.fillna(method="bfill")
print(df6)
```

```python
import numpy as np
import pandas as pd

data = {
    "이름": ["서준", "하은", "민준", "서연", np.nan, "지민"],
    "나이": [22, 28, np.nan, 31, 27, 24],
    "점수": [89, 39, 83, 90, 88, 93]
}
df = pd.DataFrame(data)

# 결측값이 존재하는 행을 아예 삭제해버림
df_drop = df.dropna()
print(df_drop, "\n")

# 결측값이 존재하는 열을 삭제
df3 = df.dropna(axis=1)
print(df3, "\n")

# na 값을 다 0으로 변경
df4 = df.fillna(0)
print(df4, "\n")

# 나이 컬럼 평균 계산 후 결측값 대체
avg_age = df["나이"].mean()
df5 = df["나이"].fillna(avg_age)
print("\n", df5, "\n")

# ffill (앞 값으로 결측값 채우기)
df6 = df.fillna(method="ffill")
print(df6, "\n")

# bfill (뒤 값으로 결측값 채우기)
df7 = df.fillna(method="bfill")
print(df7)

# 행에 대한 슬라이싱만 가능
# 열 슬라이싱은 iloc,loc을 이용해야 한다.
# ilock= 넘파이슬라이싱과 같다
# lock= 라벨을 기준으로 슬라이싱, 없다면 번호기준
```

```python
data = {
"도시": ["서울", "부산", "광주", "대구", np.nan, "춘천"],
"미세먼지": [45, 51, np.nan, 38, 49, np.nan],
"초미세먼지": [20, np.nan, 17, 18, 22, 19],
"강수량": [0.0, 2.5, 1.0, np.nan, 3.1, 0.0]
}
df = pd.DataFrame(data)
```

```python
print(df,"\n")
#1
print(f"미세먼지 평균: {df["미세먼지"].mean()}\n중앙값: {df["미세먼지"].median()}")
#2
print(f"초미세먼지 최댓값: {df["초미세먼지"].max()}\n최솟값: {df["초미세먼지"].min()}")
#3
print(f"각 컬럼별 결측값 개수: \n{df.isnull().sum()}")
#4
df_drop=df.dropna()
print(f"결측값 없는 행의 초미세먼지 평균: {df_drop["초미세먼지"].mean()}")
#5
df_fill=df.fillna(0)
print(f"미세먼지 합: {df_fill["미세먼지"].sum()}\n초미세먼지 합:{df_fill["초미세먼지"].sum()}")
#6
df_avg = df["미세먼지"].fillna(df["미세먼지"].mean())
print(f"평균으로 채운 뒤 표준편차 미세먼지: {np.std(df_avg)}")
```
